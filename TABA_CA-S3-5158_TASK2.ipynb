```python
import random


HANGMANPICS = [
    '''
      +---+
      |   |
          |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\\  |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\\  |
     /    |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\\  |
     / \\  |
          |
    ========='''
]


word_list = ["python", "code", "alpha", "codealpha"]
chosen_word = random.choice(word_list)
blanks = ["_"] * len(chosen_word)
attempts = 0  

print("Welcome to Hangman!")
print(HANGMANPICS[attempts])
print(" ".join(blanks))

# Game loop
while attempts < 3:
    guess = input("Guess a letter: ").lower()

    if guess in chosen_word:
        for i in range(len(chosen_word)):
            if chosen_word[i] == guess:
                blanks[i] = guess
        print("Good guess!")
    else:
        attempts += 1
        print(f"Incorrect! Attempts left: {3 - attempts}")

    
    print(HANGMANPICS[attempts])
    print(" ".join(blanks))

    if "_" not in blanks:
        print("Congratulations! You've won!")
        break
else:
    print(f"Game over! The word was: {chosen_word}")
```
